```python
import numpy as np
import re
```

```python
motion_params_path = '../../resultsABIDE_verifyCode/Preprocess_Datasink/motion_params_paths/motion_params_file_list.npy'
```

```python
motion_params_path_list = np.load(motion_params_path)
```

```python
motion_params_path_list
```

```python
import sys
sys.path.insert(0, '../tqdm/')

from tqdm import tqdm

outlier_count = 0
outliers_count_list = []
outlier_sub_id_list = []
outliers_ratio_list = []

for par in tqdm(motion_params_path_list):
#     print(re.findall('.+(\d{7}).+',par))
    sub_id = re.findall('.+(\d{7}).+',par)
    volume_count = 0
    for volume in np.genfromtxt(par):
        volume_count += 1
        x,y,z,roll,pitch,yaw = volume
        if (np.abs(np.array([x,y,z])) > 0.2).any() or (np.abs(np.array([roll,pitch,yaw])) > 10 ).any():
            outlier_count += 1
        outlier_ratio = outlier_count/volume_count
    outliers_count_list.append(outlier_count)
    outliers_ratio_list.append(outlier_ratio)
    outlier_sub_id_list.append(sub_id[0])
    outlier_count = 0
       
        
    
    
```
